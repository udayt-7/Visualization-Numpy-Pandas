```python
#Question 6
#Manually add a column ‘zone’ (North, South, East and West). Southern states are Kerala, TN, Karnataka, AP and TS. Likewise fill zone to each state (manually).
ds['States/UTs']

Zone = []

# For each row in the column,
for row in ds['States/UTs']:
    
#     print(row)
#     if row in ['Telangana' or 'Goa' or 'Andhra Pradesh'or'Maharashtra' or 'Karnataka'or 'Tamil Nadu' or 'Puducherry' or 'Lakshadweep' or 'Andaman and Nicobar Islands']:
        
#         Zone.append('South')
    
    
    if row in ['Telangana','Kerala','Goa','Andhra Pradesh','Maharashtra','Karnataka','Tamil Nadu','Puducherry','Lakshadweep','Andaman and Nicobar Islands']:
        
        Zone.append('South')
    
            
    elif row in ['Jammu and Kashmir','Himachal Pradesh','Uttar Pradesh','Delhi','Madhya Pradesh','Uttarakhand','Dadra and Nagar Haveli','Chandigarh']:
        
        Zone.append('North')
    
    elif row in ['Tripura','Sikkim','Odisha','Chhattisgarh','West Bengal','Bihar','Jharkhand','Assam','Arunachal Pradesh','Manipur','Mizoram','Nagaland','Meghalaya']:
       
        Zone.append('East')
    
    elif row in ['Punjab','Haryana','Gujarat','Daman and Diu','Rajasthan']:
        
        Zone.append('West')
   
    else:
        
        Zone.append('Failed')
        

ds['Zone'] = Zone

#Question 7
#Find which Zone has max and min accidents due to alcohol and drug related accidents. (2008 to 2016)

ds.set_index('States/UTs')

p = ds[['States/UTs','AAD 2008','AAD 2009','AAD 2010','AAD 2011','AAD 2012',' AAD 2013','AAD 2014','AAD 2015','AAD 2016','Zone']]
p = p.sort_values(['AAD 2008','AAD 2009','AAD 2010','AAD 2011','AAD 2012',' AAD 2013','AAD 2014','AAD 2015','AAD 2016'], ascending = False)

p = p.fillna(0)
p = p.groupby('Zone')

qw=p.sum(axis=1)
r=qw.sum(axis=1)

minzone = r[r==r.min()]
maxzone = r[r==r.max()]

maxzone
minzone
```
